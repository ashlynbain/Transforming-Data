## Splittng the data

```python
predf = predf.split('\n\n')
df2 = predf[2:3]
predf = predf[3:]
fef = [] 
for i in range(len(predf)):
    predf[i] = predf[i].split('\n')[2:]
    fef += predf[i]
```

```python
df_ = StringIO('\n'.join(fef))
df = pd.read_csv(df_, header = None)
df = df.drop(columns=[0,1,12,13])
```
